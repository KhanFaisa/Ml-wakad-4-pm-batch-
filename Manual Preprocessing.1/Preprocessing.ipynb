#### Step 4: Missing Value Treatment 
1. Categorical values replace by mode(most frequntly occuring)
2. Continousvalue replace by mean o median 

```python
df['AirBags'].value_counts()
```

```python
df['AirBags'].mode()[0]
```

```python
df['Price'].mean()
```

```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing Values Replaced in dataframe')
```

## Weight ~ Remaining Features 

## Step 5: Seprate X and Y (Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

##### Preprocessing is always done on X features 

```python
X_cat = X.select_dtypes(include='object')
X_con = X.select_dtypes(exclude='object')
```

```python
X_cat
```

```python
X_con
```

### Categorical Features Apply One Hot Encoding 


```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',
                    sparse_output=False).set_output(transform='pandas')
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```

```python
X_cat_pre.to_csv('Ohe.csv', index=False)
```

### Continous Features Apply Standard Scaler 

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_con.head()
```

```python
scaler = StandardScaler().set_output(transform='pandas')
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```

## Combine Categorical And Continous Features 

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

```python
X_pre.to_csv('PreprocessedCars.csv', index=False)
```
